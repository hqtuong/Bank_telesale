```python
#Xóa bỏ những dòng trống
df=df.dropna()
```

```python
#Xóa bỏ những cột không dùng đến (campaign, pdays, previous, poutcome)
df1=df.drop(['campaign', 'pdays', 'previous', 'poutcome'],axis=1,inplace=True)
```

```python
#Tạo cột id để định danh cho từng khách hàng
df['id'] = df.reset_index().index +1
```

```python
#Tạo cột Employment cách 1
job_mapping = {
    'admin.': 'Employed',
    'blue-collar': 'Employed',
    'entrepreneur': 'Employed',
    'housemaid': 'Employed',
    'management': 'Employed',
    'retired': 'Unemployed',
    'self-employed': 'Employed',
    'services': 'Employed',
    'student': 'Unemployed',
    'technician': 'Employed',
    'unemployed': 'Unemployed',
    'unknown': 'Unemployed'
}

df['Employment'] = df['job'].map(job_mapping)
```

```python
#Tạo cột Employment cách 2
import numpy as np

df['Employment'] = np.where(df['job'].isin(['admin.','blue-collar','housemaid','management','services','technician']), 'Employed', 'Other')
```

```python
df['Eligibility']=None
```

```python
#Credit card secure
df_card_secured=df[(df['balance'] >=1000) & (df['age']>=18)]#.sort_values(['contact','duration'],ascending=[True,False]).reset_index()
```

```python
df_card_secured
```

```python
#Credit card unsecured cách 1
df[(df['age']<=60) & (df['balance']>=4000) & (df['default']=='no') & (df['job'] != 'unemployed') & (df['job'] != 'unknown') & (df['job'] != 'student') & (df['job'] != 'blue-collar') & (df['job'] != 'housemaid')].sort_values(['contact','duration'],ascending=[True,False]).reset_index()
```

```python
#Credit card unsecured cách 2
df_card_unsecured=df[((df['age']<=60) & (df['age']>=21) & (df['Employment']=='Employed') & (df['default']=='no') & (~df['job'].isin(['blue-collar','housemaid']))) | ((df['age']<=60) & (df['age']>=21)& (df['balance']>=1080) & (df['job'].isin(['self-employed','entrepreneur','retired']) & (df['default']=='no')))]
```

```python
df_card_unsecured
```

```python
#personal loan
df_loan=df[(df['age']>=24) & (df['age']<=60) & (df['Employment']=='Employed') & (df['loan']=='no') & (df['default']=='no') & (~df['job'].isin(['blue-collar','housemaid']))].sort_values(['contact','month','duration'],ascending=[True,False,False])
```

```python
df_loan
```

```python
#map secured vào bảng tổng
df['Eligibility']=df['id'].isin(df_card_secured['id']).map({True: 'Secured',False:''})
```

```python
#map unsecured vào bảng tổng
df['Eligibility2']=df['id'].isin(df_card_unsecured['id']).map({True: 'Unsecured',False:''})
```

```python
#map personal loan vào bảng tổng
df['Eligibility3']=df['id'].isin(df_loan['id']).map({True: 'Loan',False:''})
```

```python
df=df[['id','age','job','Employment','marital','education','default','balance','housing','loan','contact','day','month','duration','y','Eligibility','Eligibility2','Eligibility3']]
```

```python
df=df.sort_values(['Eligibility3','Eligibility2','Eligibility','contact','month','duration'],ascending=[False,False,False,True,False,False])
```

```python
month_order = ['jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec']
```

```python
df['month'] = pd.Categorical(df['month'], month_order)
```
